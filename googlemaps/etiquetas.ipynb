```python
data = pd.read_csv("locales.csv")
```

```python
def type_of_establishment(name):
    name = name.lower()
    
    # Specific categories with multiple keywords
    if "hotel" in name or "lodging" in name:
        return "Hotel"
    elif "restaurant" in name or "restaurante" in name or "meal_takeaway" in name or "cafe" in name or "food" in name:
        return "Restaurant"
    elif "bar" in name:
        return "Bar"
    elif "car_rental" in name or "car_dealer" in name or "car_repair" in name:
        return "Automotive Services"
    elif "hospital" in name or "doctor" in name or "health" in name or "physiotherapist" in name:
        return "Health Services"
    elif "hair_care" in name or "beauty_salon" in name:
        return "Beauty Services"
    elif "school" in name or "university" in name:
        return "Education"
    elif "church" in name or "place_of_worship" in name:
        return "Religious Institution"
    elif "store" in name or "supermarket" in name or "book_store" in name or "electronics_store" in name or "clothing_store" in name or "home_goods_store" in name:
        return "Retail"
    elif "spa" in name:
        return "Spa"
    elif "parking" in name:
        return "Parking"
    elif "tourist_attraction" in name or "museum" in name or "art_gallery" in name or "library" in name:
        return "Tourist Attraction"
    elif "gas_station" in name or "convenience_store" in name:
        return "Gas Station"
    elif "finance" in name or "bank" in name or "atm" in name:
        return "Financial Services"
    elif "veterinary_care" in name:
        return "Veterinary Services"
    elif "laundry" in name:
        return "Laundry"
    elif "general_contractor" in name or "painter" in name:
        return "Contractor Services"
    elif "cemetery" in name:
        return "Cemetery"
    elif "movie_theater" in name:
        return "Entertainment"
    elif "local_government_office" in name:
        return "Government Services"
    elif "storage" in name:
        return "Storage Services"
    
    # Fallback for unclassified establishments
    return "Other"
```

```python
data["tipo"] = data["nombre"].apply(type_of_establishment)
```
